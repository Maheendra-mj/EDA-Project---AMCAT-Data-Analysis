```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame called 'df'
# If not, you'd load it like this:
# df = pd.read_csv('your_data.csv')

# Select numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# 1. Scatter plot matrix
plt.figure(figsize=(20, 20))
sns.pairplot(df[numerical_columns], diag_kind='kde')
plt.tight_layout()
plt.savefig('scatter_plot_matrix.png')
plt.close()

# 2. Correlation heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.close()

# 3. Hexbin plot for salary vs other variables
salary_vars = ['Salary', 'collegeGPA', 'English', 'Logical', 'Quant', 'Domain']
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.flatten()

for i, var in enumerate(salary_vars[1:]):
    plt.sca(axes[i])
    plt.hexbin(df['Salary'], df[var], gridsize=20, cmap='viridis')
    plt.colorbar(label='Count')
    plt.xlabel('Salary')
    plt.ylabel(var)
    plt.title(f'Hexbin Plot: Salary vs {var}')

plt.tight_layout()
plt.savefig('hexbin_plots.png')
plt.close()

# 4. Pair plot for personality traits
personality_traits = ['conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']
sns.pairplot(df[personality_traits], diag_kind='kde')
plt.suptitle('Pair Plot of Personality Traits', y=1.02)
plt.tight_layout()
plt.savefig('personality_traits_pairplot.png')
plt.close()

print("All plots have been saved as PNG files.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Filter for CS graduates (using the filter we created earlier)
cs_specializations = [
    'computer engineering', 'computer science & engineering', 'computer application',
    'computer science and technology', 'computer and communication engineering',
    'computer networking', 'computer science'
]
cs_filter = df['Specialization'].str.lower().str.contains('|'.join(cs_specializations))
cs_graduates = df[cs_filter]

# Convert salary to lakhs (assuming it's in rupees)
cs_graduates['Salary_Lakhs'] = cs_graduates['Salary'] / 100000

# Calculate statistics
mean_salary = cs_graduates['Salary_Lakhs'].mean()
median_salary = cs_graduates['Salary_Lakhs'].median()
pct_above_3_lakhs = (cs_graduates['Salary_Lakhs'] > 3).mean() * 100

print(f"Mean salary for CS graduates: {mean_salary:.2f} lakhs")
print(f"Median salary for CS graduates: {median_salary:.2f} lakhs")
print(f"Percentage of CS graduates earning more than 3 lakhs: {pct_above_3_lakhs:.2f}%")

# Visualize the salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(cs_graduates['Salary_Lakhs'], kde=True)
plt.title('Salary Distribution of CS Graduates')
plt.xlabel('Salary (in lakhs)')
plt.ylabel('Count')
plt.axvline(3, color='r', linestyle='--', label='3 lakhs threshold')
plt.legend()
plt.savefig('cs_salary_distribution.png')
plt.close()

# Part 2: Analyzing relationship between gender and specialization

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2}")
print(f"p-value: {p_value}")

# Visualize the relationship
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Specialization', hue='Gender')
plt.title('Distribution of Specializations by Gender')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig('gender_specialization_distribution.png')
plt.close()

print("\nPlots have been saved as PNG files.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# 1. Impact of College Tier on Career Outcomes

def analyze_college_tier_impact():
    # Calculate average salary by college tier
    tier_salary = df.groupby('CollegeTier')['Salary'].mean().sort_values(ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=tier_salary.index, y=tier_salary.values)
    plt.title('Average Salary by College Tier')
    plt.xlabel('College Tier')
    plt.ylabel('Average Salary')
    plt.savefig('salary_by_college_tier.png')
    plt.close()

    # Analyze job designations by college tier
    designation_by_tier = pd.crosstab(df['CollegeTier'], df['Designation'], normalize='index')
    plt.figure(figsize=(12, 8))
    designation_by_tier.plot(kind='bar', stacked=True)
    plt.title('Job Designations by College Tier')
    plt.xlabel('College Tier')
    plt.ylabel('Proportion')
    plt.legend(title='Designation', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig('designations_by_college_tier.png')
    plt.close()

    print("College Tier Analysis: Plots saved as 'salary_by_college_tier.png' and 'designations_by_college_tier.png'")

# 3. Influence of Personality Traits on Career Choices

def analyze_personality_traits():
    personality_traits = ['conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']

    # Correlation between personality traits and salary
    corr_salary = df[personality_traits + ['Salary']].corr()['Salary'].drop('Salary')

    plt.figure(figsize=(10, 6))
    sns.heatmap(corr_salary.to_frame().transpose(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation between Personality Traits and Salary')
    plt.tight_layout()
    plt.savefig('personality_salary_correlation.png')
    plt.close()

    # Average personality traits by job designation
    traits_by_designation = df.groupby('Designation')[personality_traits].mean()

    plt.figure(figsize=(12, 8))
    sns.heatmap(traits_by_designation, annot=True, cmap='YlGnBu')
    plt.title('Average Personality Traits by Job Designation')
    plt.tight_layout()
    plt.savefig('personality_traits_by_designation.png')
    plt.close()

    print("Personality Traits Analysis: Plots saved as 'personality_salary_correlation.png' and 'personality_traits_by_designation.png'")

# Run the analyses
analyze_college_tier_impact()
analyze_personality_traits()

print("All analyses complete. Please check the saved PNG files for visualizations.")
```
